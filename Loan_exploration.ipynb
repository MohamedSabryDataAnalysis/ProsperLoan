```python
#to check on the dates of the loan originnation first will convert it to datetime format

df['LoanOriginationDate']= pd.to_datetime(df['LoanOriginationDate'])
df.info()
```

```python
#Checking distribution of loans origination date across Years, Months and Day of month. check if we can find a trend here.

plt.figure(figsize=(10,14))

plt.subplot(3, 1, 1) 
year = df.groupby(df['LoanOriginationDate'].dt.year)['LoanOriginationDate'].count()
ax1 = year.plot.barh()
ax1.invert_yaxis()
ax1.title.set_text('Count Of Loans Per Origination Year')


plt.subplot(3, 1, 2) 
month = df.groupby(df['LoanOriginationDate'].dt.month)['LoanOriginationDate'].count()
ax2 = month.plot.barh()
ax2.invert_yaxis()
ax2.title.set_text('Count Of Loans Per Origination Month')

plt.figure(figsize=(10,30))

plt.subplot(3, 1, 3) 
days = df.groupby(df['LoanOriginationDate'].dt.day)['LoanOriginationDate'].count()
ax3 = days.plot.barh()
ax3.invert_yaxis()
ax3.title.set_text('Count Of Loans Per Origination Day of Month')
```

nothing of intrest in the day and month plots however in the years plot we can see that 2013 was the highest and 2009 was the lowest. was that related to change in intrest ratrs?

```python
#create column for loan origination  year
df['LoanOriginationYear'] = df['LoanOriginationDate'].dt.year

#Getting yearly means of Borrower APR in order to find if there is any abnormality
BorrowerAPRYearlyMean = df.groupby('LoanOriginationYear')['BorrowerAPR'].mean()
BorrowerAPRYearlyMean.dropna(inplace = True)
```

```python
#plotting the yearly means of APR
ax = BorrowerAPRYearlyMean.plot.barh()
ax.invert_yaxis()
ax.title.set_text('Yearly Means Of APR')
```

No clear relation between how count of loans and APR. Even when we check the highest year in number of loans (2013) and the lowest year (2009) vs the yearly mean of APR we find that the mean for both years is almost the same

```python
#what is the most popular term of loans in months
sorted_counts = df['Term'].value_counts();
ax = plt.pie(sorted_counts, labels = sorted_counts.index, startangle = 90, counterclock = False, wedgeprops = {'width' : 0.3});
plt.axis('square');
plt.title('Distribution Of Loan Term In Months');
```

By far the most papular term is 36 month and the least is 12 month.

```python
#checking the BorrowerRate and BorrowerAPR. Usually they are very close with the apr larger than rate as
#the apr equals the rate plus any charges
plt.figure(figsize=(15,6))

plt.subplot(1, 2, 1) 
bins = np.arange(0, df.BorrowerAPR.max()+0.05, 0.01)
plt.hist(data = df , x = 'BorrowerRate', bins=bins);
plt.title('Borrower Rate');

plt.subplot(1, 2, 2)
bins = np.arange(0, df.BorrowerAPR.max()+0.05, 0.01)
plt.hist(data = df , x = 'BorrowerAPR', bins=bins);
plt.title('Borrower APR');
```

As expected we can see the distribution is almost the same for APR and rate with the APR slightly higher due to the added charges. Distributions for both are multimodal

```python
#EmploymentStatusDuration vs LoanOriginalAmount

sb.regplot( data = df, x = 'EmploymentStatusDuration', y = 'LoanOriginalAmount', truncate=False, x_jitter=0.3, scatter_kws={'alpha':.1});
plt.title("Employment StatusDuration vs Loan Original Amount");
plt.xlabel('Employment Status Duration');
plt.ylabel('Loan Original Amount');
```

Seems that when person has more employment duration he can get loan with higher value. This is might be due to higher salaries thus ability to pay higher instalments

```python
#BorrowerAPR vs DebtToIncomeRatio
sb.regplot(data = df, x = 'BorrowerAPR', y = 'DebtToIncomeRatio', scatter_kws={'alpha':.005});
plt.xlabel('Borrower APR')
plt.ylabel('Debt To Income Ratio')
plt.ylim(0, 1);
plt.title('BorrowerAPR vs Debt To Income Ratio');
```

```python
# Checking the relation between Stated Monthly Income and whether or not it’s verifiable and the impact of the two variables on the Prosper Rating

sb.pointplot(data = df, x = 'ProsperRating (Alpha)', y = 'StatedMonthlyIncome', hue = 'IncomeVerifiable', linestyles = '', dodge = 0.2);
```

As we can see the higher verifiable monthly income the better the prosper rating. And the lower unverifiable monthly income the worse proper rating 

```python
#The impact of ProsperRating on BorrowerAPR and LoanOriginalAmount

g=sb.FacetGrid(data=df, col='ProsperRating (Alpha)',margin_titles = True, height = 8.27/3 , col_wrap=3, aspect = (14.70/3)/(8.27/3)
 )
g.map(sb.regplot, 'LoanOriginalAmount', 'BorrowerAPR', x_jitter=0.04, scatter_kws={'alpha':0.1}, line_kws={"color":"red"});
```

As we can see that the better rating the lower APR and the worse rating means higher APR, Which is totally understandable as lower rating means higher risk. However the interesting part is related to Loan Original Amount. We can see that in case of a good rating (AA or A) that higher loan amount means higher APR. While on the lower ratings (From B to HR) the higher loan amount the lower APR. This may be due to if higher rating want higher amount they will not complain from the higher APR on the other hand maybe lower rating will not want to get high amount loans so lower APR is for encouraging them to take larger loans.

We see here as expected that the same income range gets lower APR in case it was verifiable. Please note that this is with the exception of income range 0 and not displayed which are not really an income range
